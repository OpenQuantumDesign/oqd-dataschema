```python
import pathlib

import numpy as np
from rich.pretty import pprint

from oqd_dataschema.base import Dataset, GroupBase, GroupRegistry
from oqd_dataschema.datastore import Datastore
from oqd_dataschema.groups import (
    SinaraRawDataGroup,
)
```

```python
class YourCustomGroup(GroupBase):
    """
    Here we define a custom Group, which is automatically added at runtime to the GroupRegistry.
    """

    array: Dataset
```

```python
GroupRegistry.groups
```

```python
dtype = "int32"
data = np.ones([10, 10]).astype(dtype)
custom_group = YourCustomGroup(array=Dataset(data=data))

dtype = "int32"
data = np.ones([10, 10]).astype(dtype)
default_group = SinaraRawDataGroup(camera_images=Dataset(data=data))
```

```python
filepath = pathlib.Path("test.h5")

datastore = Datastore(
    groups={"default_group": default_group, "custom_group": custom_group}
)
pprint(datastore)
datastore.model_dump_hdf5(filepath)
```

```python
parse = Datastore.model_validate_hdf5(filepath)
pprint(parse)
```
