```python
import pathlib

import numpy as np
from rich.pretty import pprint

from oqd_dataschema import Dataset, Datastore, GroupBase, GroupRegistry, condataset
```

```python
class YourCustomGroup(GroupBase):
    """
    Here we define a custom Group, which is automatically added at runtime to the GroupRegistry.
    """

    array: condataset(shape_constraint=(None, 10))  # type: ignore
```

```python
GroupRegistry.groups
```

```python
from datetime import datetime

dtype = "int32"
data = np.ones([10, 10]).astype(dtype)
custom_group = YourCustomGroup(
    array=Dataset(data=data), attrs={"created_at": f"{datetime.now()}"}
)
```

```python
filepath = pathlib.Path("test.h5")

datastore = Datastore(groups={"custom_group": custom_group})
pprint(datastore)
datastore.model_dump_hdf5(filepath)
```

```python
parse = Datastore.model_validate_hdf5(filepath)
pprint(parse)
```

```python
from typing import Dict

from oqd_dataschema import CastDataset


class A(GroupBase):
    data: Dict[str, CastDataset]
```

```python
filepath = pathlib.Path("test.h5")

datastore = Datastore(
    groups={
        "A": A(
            data={"x": Dataset(data=np.random.rand(10), attrs={"type": "mytype"})},
            attrs={"created_at": f"{datetime.now()}"},
        )
    }
)
pprint(datastore)
datastore.model_dump_hdf5(filepath)
```

```python
parse = Datastore.model_validate_hdf5(filepath)
pprint(parse)
```
